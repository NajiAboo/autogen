```python
import autogen
```

```python
config_list = autogen.config_list_from_json(
    env_or_file ="/content/OAI_CONFIG_LIST",
    filter_dict={
        "model" : ["gpt-3.5-turbo-16k"]
    }
)
```

```python
llm_config = {
    "temperature": 0,
    "seed": 1,
    "timeout": 200,
    "config_list": config_list
}
```

```python
assistant = autogen.AssistantAgent(
    name="assistant",
    llm_config=llm_config
)
```

```python
user_proxy = autogen.UserProxyAgent(
    name="userproxy",
    system_message="user proxy",
    human_input_mode="TERMINATE",
    is_termination_msg= lambda x: x.get("content","").rstrip().endswith("TERMINATE"),
    max_consecutive_auto_reply=3,
    code_execution_config={
        "work_dir" : "program",
        "use_docker" : False

    }
)
```

```python
user_proxy.initiate_chat(assistant, message="Write a proram that will check the current date and will compare year-to-date gain from META and TESLA")
```
